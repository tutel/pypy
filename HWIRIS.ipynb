```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.datasets import load_iris
```

```python
I = load_iris()
```

```python
x = I.data
y = I.target
```

```python
x[:5]
```

```python
Y = y
```

```python
X = x[:,:2]
```

```python
len(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X, Y, test_size = 0.6, random_state = 0)
```

```python
plt.scatter(X[:,0],X[:,1],c = Y)
```

```python
cl = SVC(gamma = "auto")
```

```python
cl.fit(x_train,y_train)
```

```python
yp = cl.predict(x_test)
```

```python
yd = yp - y_test
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c = yd,cmap = "Paired")
```

```python
x0 = np.arange(4,8.2,0.02)
y0 = np.arange(1.8,4.7,0.02)
xm, ym = np.meshgrid(x0,y0)

P = np.c_[xm.ravel(),ym.ravel()]
z = cl.predict(P)
Z = z.reshape(xm.shape)

plt.contourf(xm,ym,Z, cmap = 'Paired',alpha = 0.2)
plt.scatter(x_test[:,0]+0.33,x_test[:,1],c = y_test,cmap = 'Paired')
```
