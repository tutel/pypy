```python
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
import numpy as np
```

```python
def U():
    x = mpimg.imread('中秋.jpg')
    plt.imshow(x)
    plt.axis('off');
    
def T():
    x = mpimg.imread('國慶.png')
    plt.imshow(x)
    plt.axis('off');
    
```

```python
x="r"
while "掰掰" not in x:
    x = input("今天是甚麼節日?\n")
    if x=="中秋":
        U()
        break
    elif x=="國慶":
        T()
        break
    elif "掰" in x:
        print("掰掰")
        break
    else :
        print("我找不到圖片")
```
