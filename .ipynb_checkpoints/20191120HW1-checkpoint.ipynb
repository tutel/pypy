#  三角函數在同一張圖畫出來

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def sin():
    x = np.linspace(-5,5,200)
    y = np.sin(x)
    plt.plot(x,y)
def cos():
    x = np.linspace(-5,5,200)
    y = np.cos(x)
    plt.plot(x,y)
def tan():
    x = np.linspace(-5,5,200)
    y = np.tan(x)
    y[y>5] = np.inf
    y[y<-5] = -np.inf
    plt.plot(x,y)
def cot():
    x = np.linspace(-5,5,200)
    y = 1/np.tan(x)
    y[y>5] = np.inf
    y[y<-5] = -np.inf
    plt.plot(x,y)
def sec():
    x = np.linspace(-5,5,200)
    y = 1/np.cos(x)
    y[y>5] = np.inf
    y[y<-5] = -np.inf
    plt.plot(x,y)
def csc():
    x = np.linspace(-5,5,200)
    y = 1/np.sin(x)
    y[y>5] = np.inf
    y[y<-5] = -np.inf
    plt.plot(x,y)
```
